#### Train YOLO26 model to detect damages

```python
from ultralytics import YOLO

model = YOLO("yolo26m-seg.pt")

model.train(
    data="/content/CSK_Dataset/CSK_dataset.yaml",
    epochs=150,
    patience=20,
    imgsz=1280,
    batch=24,
    workers=8,
    plots=True,
    save_period=10,
)
```

#### Resume training in case of crash (just change the model path)

```python
# Resume training , just updated the save directory to be on the drive (weight automatically get saved to the drive)

model = YOLO("/content/runs/segment/train-2/weights/last.pt")

model.train(
    data="/content/CSK_Dataset/CSK_dataset.yaml",
    epochs=130,
    patience=20,
    imgsz=1280,
    batch=24,
    workers=8,
    plots=True,
    save_period=5,
    save_dir='/content/drive/MyDrive/Mini Vision Datas',
    name='CSK_model'
)


#OR resume training using.
model.train(
    data="/content/CSK_Dataset/CSK_dataset.yaml",
    resume=True,
)
```
